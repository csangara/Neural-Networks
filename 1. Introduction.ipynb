## What is a Neural Network?

Neural networks are one of the many models in **machine learning**, a field of study that, as described by Arthur Samuel,  "gives computers the ability to learn without being explicitly programmed." Machine learning has been gaining tremendous popularity over the last few decades, and it can be used in predicting, classifying, and clustering data.

There are two types of machine learning: supervised learning and unsupervised learning. **Supervised learning** refers to algorithms in which the "right" answers are given. The data given to the machine is already labeled, an example being a dataset of house sizes that is already labeled with the corresponding prices. We can then use that information to predict the price of house sizes not in the dataset. In  **unsupervised learning**, however, there is no labeled 'answer.' The algorithm must try to make sense of the data by itself. In this tutorial, we will focus on supervised learning.

Applications of machine learning range from the mundane e-mail spam classification and Netflix movie recommendations to the remarkable image recognition and autonomous driving. Neural networks are especially useful in classification problems - for instance, given a tumor size, the algorithm must predict whether the tumor is benign or malignant based on past information. Nowadays, neural networks are considered the state-of-the-art algorithms in solving speech and image recognition problems.

## Biological Motivation

Neural networks, as its name suggests, are inspired by the human brain. The **neuron** is the main computational unit in the human brain (its picture is shown below).

Neurons use pulses of electricity to send messages to one another. The dendrite of a neuron receives a signal, processes it, and fires it to the axon terminal if a **threshold** value is exceeded.

Artificial neurons in a neural network work in a similar way: it takes an input data, performs simple mathematical operations on it, and outputs the processed data.

## How It Works

Each neuron has **weight** values associated to it, and these constrain how the input data is related to output data. You can visualize the weights as the edge connecting one neuron to another.

What an individual neuron does is take the input, multiply it with the weight(s), and pass the processed data (also called the **activation**) along to the next neuron.  The neuron outputs a 1 if the activation value exceeds the threshold, and 0 if it does not.

In the tumor classification problem mentioned previously, an output of 0 could mean that the tumor is benign, and 1 could mean that the tumor is malignant.

Let's look at the simple network below. The activation being sent to the output layer is computed by:

$$activation = (input1 \times weight1) + (input2 \times weight2) $$

```python
Image('activation.png')
```

In a network of several layers, we can have outputs that model even the most complicated functions by varying the values of the weights and thresholds. You might be asking: *how would you decide the values of the weights (and threshold) in the first place?*

These values are obtained during the training period. That is, we feed **training data** with inputs $x$ and the correct output $y$ to our network. This could be points on a graph, images, or even audio clips. Through an iterative process, the weights are going to converge to a local optimum, and we can then use the network to classify new data.

## Application - Handwriting Recognition

One of the earliest applications of neural networks is handwriting recognition, in which the algorithm is able to classify handwritten digits from 0 to 9. It has an accuracy of over 99%, and therefore should not be surprising that post offices and banks use such algorithms to recognize addresses and process cheques.

What would our network look like for this problem?

In general, the number of input nodes correspond to the number of *features* of the data. For handwriting recognition, the [MNIST database](http://yann.lecun.com/exdb/mnist/) is most commonly used. These are 20x20 grayscale images of a handwritten digit, with the correct classification given. So, one way to do it is to have one node correspond to one pixel, giving a total of 400 input nodes. Samples of images from the dataset are shown below.

```python
Image('digits.png')
```

The number of output nodes correspond to the number of classes. Since we must classify numbers from 0 to 9, there are 10 classes in total, and we would need 10 output nodes.

The number of hidden layers and the nodes in hidden layers, on the other hand, is more of a trial-and-error process. There is no definite way of determining how many you will need, but it is a good idea to just start off with one hidden layer when first implementing your network. (Networks with many hidden layers - also called *deep neural networks* - come with LOTS of problems!)

So, here is what our network might look like:

```python
Image('digitsnn.png')
```

If the network thinks the digit is a '0', the first output node will return a 1 and the rest of the output nodes will return a 0. If the network thinks the digit is a '1', the second output node will return a 1, and the rest will return a 0. The same can be said for the other digits.

Note again that the number of nodes in the hidden layer doesn't have to be 10 - usually, it is determined through trial-and-error.

That's the basic idea of neural networks. In the next section, we will cover single-layer networks - McCulloch Pitts and perceptrons - which do not contain a hidden layer.
